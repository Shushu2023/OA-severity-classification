```python
#no need to run optional
#to fix the relative path between windows and linux
# Read the file
with open('/content/OA-severity-classification/data_pipeline/preprocessing.py', 'r') as f:
    content = f.read()

# Check if fix is already there
if "replace('\\\\', '/')" in content or 'replace(\'\\\\\', \'/\')' in content:
    print("Fix already present")
else:
    # Apply the fix
    old = "full_path = os.path.join(self.base_dir, relative_path)"
    new = "relative_path_fixed = relative_path.replace('\\\\', '/')\n        full_path = os.path.join(self.base_dir, relative_path_fixed)"

    if old in content:
        content = content.replace(old, new)
        with open('/content/OA-severity-classification/data_pipeline/preprocessing.py', 'w') as f:
            f.write(content)
        print("Fix applied successfully")
    else:
        print("Could not find the line to replace — check manually")
```
