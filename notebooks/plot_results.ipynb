```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
train_dir = '../reports/training_history_resnet.json'
test_dir = '../reports/test_results_resnet.json'
with open(train_dir, 'r') as file:
    train = json.load(file)
with open(test_dir, 'r') as file:
    test = json.load(file)

fig,ax = plt.subplots(nrows=1,ncols=3,figsize=(12,3))
x = np.linspace(1,30,30)
ax[0].plot(x,train['train_loss'],label='Train')
ax[0].plot(x,train['val_loss'],label='Validation')
ax[0].legend()
ax[0].set_title('Loss')

ax[1].plot(x,train['train_acc'],label='Train')
ax[1].plot(x,train['val_acc'],label='Validation')
ax[1].legend()
ax[1].set_title('Accuracy')

ax[2].plot(x,train['train_f1'],label='Train')
ax[2].plot(x,train['val_f1'],label='Validation')
ax[2].legend()
ax[2].set_title('F1')

fig.supxlabel('Epoch')
fig.supylabel('Score')

plt.tight_layout()
plt.show()
```
